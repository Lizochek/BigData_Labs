Проверка распределений на нормальность (критерий Шапиро-Уилка) с
использованием scipy.stats: 

```python
import scipy.stats as sts
```

```python
a = data2.loc[data2['region']=='southwest']
a=a['bmi'].astype(float)
b = data2.loc[data2['region']=='northwest']
b=b['bmi'].astype(float)
```

```python
res1 = sts.shapiro(a)
res2 = sts.shapiro(b)
print(res1, '\n', res2)
```

Тест дает p-значение выше 0.05, следовательно, выборки имеют
нормальное распределение

Проверка гомогенности дисперсии. Критерий Бартлетта – статистический
критерий, позволяющий проверять равенство дисперсий нескольких (двух и
более) выборок. Нулевая гипотеза предполагает, что рассматриваемые
выборки получены из генеральных совокупностей, обладающих одинаковыми
дисперсиями. Так же предполагается, что выборки распределены нормально.


```python
res = sts.bartlett(a,b)
print(res)
```

p-уровень превышает 0.05, следовательно, дисперсии выборок примерно
одинаковы. Можем переходить к t критерию Стьюдента.

```python
t_res = sts.ttest_ind(a,b)
print(t_res)
```

p-значение намного ниже 0.05, следовательно нулевая гипотеза
отвергается, выборки принадлежат к разным генеральным совокупностям и их
средние значения различны.
Если распределение выборки очень сильно отличается от нормального, то
можно использовать непараметрический критерий Манна-Уитни

```python
data3 = {'Points': [1, 2, 3, 4, 5, 6],'Observed': [97, 98, 109, 95, 97, 104]} 
throws = pd.DataFrame(data3) 
throws['Expected'] = 100
throws
```

```python
import scipy.stats as st
k = len(throws.Points) - 1 # число степеней свободы
difference = throws.Observed - throws.Expected
throws['Difference'] = difference
throws['SquareDif'] = throws['Difference'] **2
throws['SquareDif/Expected'] = throws['SquareDif'] / throws.Expected
print(throws)
statistic = throws['SquareDif/Expected'].sum()
print('\nСтатистика: ', statistic)
pval = 1 - st.chi2.cdf(statistic, k)
print('Вероятность получить похожее значение или выше, исходя из полученной статистики: ', pval)
st.chisquare(throws['Observed'], throws['Expected'])
```

Основываясь на p-value делаем вывод, что распределение равномерное

```python
data4 = pd.DataFrame({'Женат': [89,17,11,43,22,1],
 'Гражданский брак': [80,22,20,35,6,4],
 'Не состоит в отношениях': [35,44,35,6,8,22]})
data4.index = ['Полный рабочий день','Частичная занятость',
               'Временно не работает','На домохозяйстве','На пенсии','Учёба']
data4
```

```python
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(data4)
alpha = 0.05
print ( "p value is " + str (p))
if p <= alpha:
  print('Переменные зависимы (отклонена гипотеза H0). Семейное положение влияет на занятость')
else :
  print('Переменные независимы (гипотеза H0 верна). Семейное положение не влияет на занятость')
```

Функция chi2_contingency () модуля scipy.stats принимает в качестве входных данных таблицу непредвиденных обстоятельств в формате 2d-массива. Он возвращает кортеж, содержащий статистику теста , p-значение , степени свободы и ожидаемую таблицу (ту, которую мы создали из вычисленных значений) в указанном порядке.
